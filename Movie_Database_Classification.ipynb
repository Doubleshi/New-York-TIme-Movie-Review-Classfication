```python
#Option 1: name as creatures, genre as label

from stop_words import get_stop_words

def title_features(title):
    # stopwords list is all lowercase so we need to match
    title = title.lower()
    words = title.split()
    keywords = [x for x in words if x not in get_stop_words('english')]
    # I tried to ignore some symbols and numbers, but it will print the whole feature_sets out as none instead
    # characters_ihate = ['(',')','-','']
    # keywords = [term for term in words if term not in keywords and not any (characters in term for characters in characters_ihate)]
    # incase some titles don't have 4 keywords, just add nothing to keep the format of 4 keywords
    # 4 is the number decided by ourselves, and can try to find out different results
    # if keywords more than 4, only gather first 4 keywords, and just ignore the rest of the keywords
    while len(keywords) < 4:
        keywords.append('None')
    return {'first_keyword': keywords[0], 'second_keyword': keywords[1], 'third_keyword': keywords[2], 'forth_keyword': keywords[3] }
```

```python
zipped_feature_and_labels = zip(movies.Title, movies.Genre)
```

```python
feature_sets = [(title_features(title),y) for title,y in zip(movies.Title,movies.Genre)]
```

```python
feature_sets
```

```python
import random
random.shuffle(feature_sets)
```

```python
import nltk
import math

print(len(feature_sets))
split_num = math.floor(len(feature_sets)*.8)
print(split_num)
```

```python
training_set = feature_sets[:split_num]
testing_set = feature_sets[split_num:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
classifier.classify(title_features("Jimmy Neutron: Boy Genius"))
```

```python
print(nltk.classify.accuracy(classifier, testing_set))
```
